```python
import pandas as pd
from IPython.display import display

link1 = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(PPP)_per_capita"
link2 = "https://en.wikipedia.org/wiki/World_Happiness_Report"
link3 = "https://en.wikipedia.org/wiki/List_of_countries_by_life_expectancy"
link4 = "https://en.wikipedia.org/wiki/List_of_sovereign_states_and_dependent_territories_by_birth_rate"


happiness = pd.read_html(link2)[5]
happiness.columns = ["Overall rank","Country/Territory","Score","GDP per capita","Social support","Healthy life expectancy","Freedom to make life choices","Generosity","Perceptions of corruption"]
happiness = happiness.drop([0,1], axis=0)
happiness
```

```python
life = pd.read_html(link3)[0]
life.rename(columns = {"Country and regions":'Country/Territory'}, inplace = True)
life
```

```python
birth = pd.read_html(link4)[1]
#Remove annoying multi-indices
birth.columns = birth.columns.map('{0[0]}|{0[1]}'.format)
birth.rename(columns = {"Country/territory|Country/territory":'Country/Territory'}, inplace = True)
birth
```

```python
gdp = pd.read_html(link1)[2]

gdp = gdp.set_index("Country/Territory")
gdp = gdp.drop(["Rank"], axis=1)

merge1 = pd.merge(gdp, happiness[["Country/Territory", "Score"]],
                  on="Country/Territory", how="left")

merge1
```

```python
merge2 = pd.merge(merge1, birth[["Country/Territory","Both sexes lifeexpectancy (HALE)"]],
                 on="Country/Territory", how="left")
merge2
```

```python
merge3 = pd.merge(merge2, birth[["Country/Territory", "CIA WF 2016|Rateâ€°"]],
                 on="Country/Territory", how="left")

merge3.columns= ["Country", "GDP(PPP) per capita", "Happiness score", "Life expectancy", "Birth rate"]
merge3
```

```python
merge3.to_html("countrydata.html")
merge3.to_csv("countrydata.csv")
merge3.to_json("countrydata.json")
```

```python
import matplotlib.pyplot as plt
jsonData = pd.read_json("countrydata.json")

fig, axs = plt.subplots(2, 3, figsize=(20, 10))
axs[0, 0].scatter(jsonData["GDP(PPP) per capita"], jsonData["Happiness score"])
axs[0, 0].set_title('GDP and Happiness score')
axs[0, 1].scatter(jsonData["GDP(PPP) per capita"], jsonData["Life expectancy"])
axs[0, 1].set_title('GDP and Life expectancy')
axs[0, 2].scatter(jsonData["GDP(PPP) per capita"], jsonData["Birth rate"])
axs[0, 2].set_title('GDP and Birth rate')
axs[1, 0].scatter(jsonData["Happiness score"], jsonData["Life expectancy"])
axs[1, 0].set_title('Happiness score and life expectancy')
axs[1, 1].scatter(jsonData["Happiness score"], jsonData["Birth rate"])
axs[1, 1].set_title('Happiness score and Birth rate')
axs[1, 2].scatter(jsonData["Life expectancy"], jsonData["Birth rate"])
axs[1, 2].set_title('Life expectancy and Birth rate')
plt.savefig("countryplots.png")

corrs = []
corrs.append(jsonData["GDP(PPP) per capita"].corr(jsonData["Happiness score"], method="pearson"))
corrs.append(jsonData["GDP(PPP) per capita"].corr(jsonData["Life expectancy"], method="pearson"))
corrs.append(jsonData["GDP(PPP) per capita"].corr(jsonData["Birth rate"], method="pearson"))
corrs.append(jsonData["Happiness score"].corr(jsonData["Life expectancy"], method="pearson"))
corrs.append(jsonData["Happiness score"].corr(jsonData["Birth rate"], method="pearson"))
corrs.append(jsonData["Life expectancy"].corr(jsonData["Birth rate"], method="pearson"))
corrs



```

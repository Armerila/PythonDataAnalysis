```python
import pandas as pd
import numpy as np

data = pd.read_csv("lastfm-matrix-germany.csv")

data1 = data[data.astype('bool').sum(axis=1)>10]
data1.drop("user", axis=1, inplace=True)
#this gives 138 columns instead of 195 for some reason
data1 = data1.replace(0,np.NaN)
data1 = data1.dropna(axis=1, thresh=35)

data1 = data1.fillna(0).astype(int)
data1
```

```python
import scipy.spatial.distance as dist


artdists = dist.squareform(dist.pdist(data1.T,"cosine"))
distDf = pd.DataFrame(artdists)

columnames = list(data1.columns.values)
distDf.columns = columnames
distDf.index = columnames


distDf
```

```python
#Drop lower half
def get_redundant_pairs(df):
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top(df, n):
    sims = df.unstack()
    labels_to_drop = get_redundant_pairs(df)
    sims = sims.drop(labels=labels_to_drop).sort_values(ascending=False)
    return sims[0:n]

#This doesn't seem right
get_top(distDf, 100)
```
